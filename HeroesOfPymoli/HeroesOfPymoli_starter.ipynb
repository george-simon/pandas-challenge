```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#CleanData
purchase_data.count()
```




## Player Count

* Display the total number of players


```python
#Count unique players > SN
#https://stackoverflow.com/questions/38309729/count-unique-values-with-pandas-per-groups
total_players=  purchase_data["SN"].nunique()
total_players

#create summary table
player_summary = pd.DataFrame({"Total Player Count": [total_players]})
player_summary

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculation for unique items
unique_items = purchase_data["Item ID"].nunique()

#Calculation for average price
average_price = purchase_data["Price"].mean()

#Calculation for number of purchases
num_purch = purchase_data["Purchase ID"].count()


#Calculation for total revenue
tot_rev = purchase_data["Price"].sum()
```

```python
#Display summary purchase data
purchase_summary = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": average_price, \
                                 "Number of Purchases": num_purch, "Total Revenue": tot_rev })

#Format dollars "${:.2f}
purchase_summary["Average Price"] = purchase_summary["Average Price"].astype(float).map("${:.2f}".format)
purchase_summary["Total Revenue"] = purchase_summary["Total Revenue"].astype(float).map("${:.2f}".format)
purchase_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#How do I calculate: Avg Total Purchase per Person??
#Hint: Multiple groupby - multiple columns

#https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.core.groupby.DataFrameGroupBy.agg.html
#Example: df.groupby('A').agg({'B': ['min', 'max'], 'C': 'sum'}
# analysis_gender = purchase_data.groupby("Gender").agg({"Purchase ID":"count","Price":["sum","mean"]})
# analysis_gender

#Purchase Count -> groupby Gender > Purchase ID.count()
purchase_count = purchase_data.groupby("Gender").count()["Purchase ID"]

#Average Purchase Price -> groupby Gender > Price.mean()
average_price = purchase_data.groupby("Gender").mean()["Price"]

#Total Purchase Price -> groupby Gender > Price.sum()
total_price = purchase_data.groupby("Gender").sum()["Price"]

```

```python
#create gender analysis dataframe
gender_analy = pd.DataFrame({"Purchase Count ": purchase_count, "Average Purchase Price": average_price,\
                             "Total Purchase Price" :total_price})

#run calculations of gender_analy df for average total purchase per person
male_df = purchase_data.loc[purchase_data["Gender"]== "Male"]
uniq_male = male_df["SN"].nunique()

fem_df = purchase_data.loc[purchase_data["Gender"]== "Female"]
uniq_fem = fem_df["SN"].nunique()

non_dis_df = total_players - (uniq_fem + uniq_male)

total_purchase_female = gender_analy.iloc[0,2]
tbf = total_purchase_female/uniq_fem 
total_purchase_male = gender_analy.iloc[1,2]
tpm = total_purchase_male/uniq_male
total_purchase_other = gender_analy.iloc[2,2]
tpn = total_purchase_other/non_dis_df

avg_tot_pp = pd.array([tbf,tpm,tpn])

gender_analysis_sum = pd.DataFrame({"Purchase Count ": purchase_count, "Average Purchase Price": average_price,\
                             "Total Purchase Price" :total_price, "Avg Total Purchase per Person": avg_tot_pp })
```

```python
#Format dollars "${:.2f}
gender_analysis_sum["Average Purchase Price"] = gender_analysis_sum["Average Purchase Price"].astype(float).map("${:.2f}".format)
gender_analysis_sum["Total Purchase Price"] = gender_analysis_sum["Total Purchase Price"].astype(float).map("${:.2f}".format)
gender_analysis_sum["Avg Total Purchase per Person"] = gender_analysis_sum["Avg Total Purchase per Person"].astype(float).map("${:.2f}".format)
gender_analysis_sum
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data.head()
```

```python
#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]

# Create the names for the 8 bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

#add column to purchase_data
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names, include_lowest=True)
```

```python
#group by new age group column
# # Create a GroupBy object based upon "View Group"
# ted_group = ted_df.groupby("View Group")

# # Find how many rows fall into each bin
age_groups = purchase_data.groupby("Age Range")
count_age = age_groups["SN"].nunique()

#percent of players in each bin
percent_players = count_age/total_players
```

```python
#Summary table
age_demo_summary = pd.DataFrame({"Total Count ": count_age, "Percentage of Players": percent_players})
```

```python
#format percentages ->"{:.2%}"
age_demo_summary["Percentage of Players"] = age_demo_summary["Percentage of Players"].astype(float).map("{:.2%}".format)
age_demo_summary
```
